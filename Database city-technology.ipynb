#### C'è PIù DI UNA VERSIONE DI IPC. COME GESTIRE? ---> passo a 4 digits

```python
np.unique(prova.ipc_version)
```

```python
'''avendo diverse versioni di ipc, alcuni codici tecnologici vanno gestiti e rivisti a mano.
La maggior parte di questo problema tuttavia e` risolvibile passando a 4 digits:
con il cambiare delle versioni vengono aggiunti altri codici relativi ad una categoria a 4 digit
(ad esempio nella versione del 2006 non e` presente il codice B31B 160, mentre e` presente in quella del 2021),
questi "codici mancanti" li aggiungo a mano nel file "ipcdict.tsv". Passando poi a 4 digit, il "rumore" relativo a
tale aggiunta viene eliminato.
Ci sono pero` anche codici a 4 digits che vengono aggiunti con il tempo. Ad esempio nella versione del 2006 e`
presente il codice F24J ---> PRODUCTION OR USE OF HEAT NOT OTHERWISE PROVIDED FOR (vediamo che e` un codice
"generico"); nella versione 2021, F24J non e` presente, ma ci sono i codici F24S, F24T ed F24V.
Ora, F24J si divide in F24J 1, F24J 2 ed F24J 3 rispettivamente "Apparatus or devices using heat produced by
exothermal chemical reactions other than by combustion", "Use of solar heat, e.g. solar heat collectors" ed
'Other production or use of heat, not derived from combustion'.
Consideriamo i primi due.
Questi codici, della versione del 2006, combaciano rispettivamente con i codici F24V 30 ed F24S (versione 2021)
(quest'ultimo codice presenta varie sottocategorie ma non e` un problema poiche` porremo tutto a 4 digit (ed
ecco che tale soluzione diviene ancora utile)). Quindi, nel file "ipcdict.tsv" sostituiremo F24J 1 ed F24J 2
(versioni del 2006) con F24V 30 ed F24S (versioni del 2021). In particolare di F24S metteremo gli ultimi due
6 digit scegliendoli a caso tra quelli esistenti (ricordiamo che concateneremo tutto a 4, quindi non e` un
problema). La stessa sostituzione andra` fatta nella variabile appln_all/prova.
Questo lavoro andra` fatto per ogni codice tecnologico a 4 digits diverso da versione a versione.

Alcuni codici vanno gestiti diversamente: F24J 3 (versione 2006) -> 'Other production or use of heat,
not derived from combustion', ha come sottoclassi F24J 3/06 'using natural heat' ed F24J 3/08 'using geothermal
heat'. Questi codici (ad 8 digit) vanno sostituiti rispettivamente con (versioni 2021) F24V ed F24T
(non importano gli ultimi 2 digit (o meglio, 4)).
Apportiamo le modifiche riguardanti gli 8 digit ora, successivamente elimineremo gli ultimi due digit e rimarremo
a 6. Se un appln_id non ha gli ultimi due degli 8 digit gli assegno successivamente F24V 99
'Subject matter not provided for in other main groups of this subclass' (la piu` generica)'''
```

```python
prova = prova.replace({'ipc_class_symbol': {'F24J   3/06':'F24V  30', 'F24J   3/08':'F24T  10'}})
```

```python
#Alcuni codici tecnologici presentano 6 digit + 2 (un esempio è G01J 1/44); devo eliminare il +2
#(nell'esempio, il /44). In questo modo però allo stesso appln_id potrebbero venire associati più volte codici
#uguali; devo correggere anche questo
prova['ipc_6'] = prova['ipc_class_symbol'].astype(str).str[0:8] #---> aggiungo una colonna con solo i primi 6 digit
prova = prova.drop_duplicates(subset=['appln_id','ipc_6']) #---> seconda correzione
```

#### Cambi relativi ai 6 digit, dopo aver eliminato gli ultimi due degli 8 totali

```python
# Cambi da fare nel file 'prova' a 6 digit
#1) F24J 1 -> F24V 30; 2) F24J 2 -> F24S 10 (ultimi due digit casuali) 3) F24J 3 -> F24V 99
prova = prova.replace({'ipc_6': {'F24J   1':'F24V  30', 'F24J   2':'F24S  10', 'F24J   3':'F24V  99'}})
```
